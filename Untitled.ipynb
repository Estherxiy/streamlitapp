```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve
import pandas as pd
from scipy.interpolate import interp1d
from numpy import argmax
from sklearn.metrics import roc_auc_score, accuracy_score, recall_score, f1_score, precision_score, confusion_matrix, brier_score_loss
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
import joblib

import warnings
warnings.filterwarnings("ignore")

df = pd.read_excel("数据集.xlsx")

X = df.iloc[:, 1:]
y = df.iloc[:, 0]

# 数据集划分
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.3, random_state=42)

classifiers = [
    ('Random Forest', RandomForestClassifier(random_state=42)),
    ('Logistic Regression', LogisticRegression(penalty='l2',C=1.0)),
    ('Decision Tree', DecisionTreeClassifier(random_state=42)),
    ('XGBoost', XGBClassifier(use_label_encoder=False, eval_metric='logloss',random_state=42)),
    ('SVM', SVC(probability=True,random_state=42))
]

def evaluate_model(name, clf, X_train, y_train, X_test, y_test):
    model = clf.fit(X_train, y_train)
    joblib.dump(model, f"{name}.pkl")
    y_proba = model.predict_proba(X_test)[:, 1]
    y_pred = model.predict(X_test)

    # 计算fpr, tpr及相应的thresholds
    fpr, tpr, thresholds = roc_curve(y_test, y_proba)

    # 计算sensitivity/recall/TPR, specificity/TNR, PPV, NPV
    sensitivity = tpr
    specificity = 1 - fpr
    ppv = sensitivity / (sensitivity + fpr+1)
    npv = specificity / (specificity + (1 - tpr))

    # 计算F1 Score
    f1_scores = 2*(ppv*sensitivity) / (ppv + sensitivity)

    # 找到最佳阈值
    ix = argmax(f1_scores)
    best_threshold = thresholds[ix]/4

    # 基于最佳阈值，得到新的预测值
    y_pred_new = [1 if i >= best_threshold else 0 for i in y_proba]

    # 基本度量
    acc = accuracy_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_proba)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)

    # confusion matrix
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()

    # specificity，brier分数，PPV和NPV
    specificity = tn / (tn+fp)
    brier = brier_score_loss(y_test, y_proba)
    ppv = tp / (tp + fp+1) # PPV就是precision
    npv = tn / (tn + fn)

    return (best_threshold, acc, auc, recall, specificity, f1, precision, ppv, npv, brier)

result_train = []
result_test = []

for name, clf in classifiers:
    train_result = evaluate_model(name, clf, X_train, y_train, X_train, y_train)
    test_result = evaluate_model(name, clf, X_train, y_train, X_test, y_test)

    result_train.append((name,) + train_result)
    result_test.append((name,) + test_result)

df_result_train = pd.DataFrame(result_train, columns=['model','best_threshold', 'accuracy', 'AUC', 'Sensitivity(Recall)', 'Specificity', 'F1_Score', 'Precision', 'PPV', 'NPV', 'Brier Score'])
df_result_test = pd.DataFrame(result_test, columns=['model','best_threshold', 'accuracy', 'AUC', 'Sensitivity(Recall)', 'Specificity', 'F1_Score', 'Precision', 'PPV', 'NPV', 'Brier Score'])

df_result_train.to_excel('model_evaluation_result_train.xlsx', index=False)
df_result_test.to_excel('model_evaluation_result_test.xlsx', index=False)
```
