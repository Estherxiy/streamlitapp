```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve
import pandas as pd
from scipy.interpolate import interp1d
from numpy import argmax
from sklearn.metrics import roc_auc_score, accuracy_score, recall_score, f1_score, precision_score, confusion_matrix, brier_score_loss
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
import joblib

import warnings
warnings.filterwarnings("ignore")

df1 = pd.read_csv("train.csv")
X_train = df1.iloc[:, :-1]
y_train = df1.iloc[:, -1]

df2 = pd.read_csv("test.csv")
X_test = df2.iloc[:, :-1]
y_test = df2.iloc[:, -1]

classifiers = [
    ('Random Forest', RandomForestClassifier(random_state=42)),
    ('Logistic Regression', LogisticRegression(penalty='l2',C=1.0)),
    ('Decision Tree', DecisionTreeClassifier(random_state=42)),
    ('XGBoost', XGBClassifier(use_label_encoder=False, eval_metric='logloss',random_state=42)),
    ('SVM', SVC(probability=True,random_state=42))
]

def evaluate_model(name, clf, X_train, y_train, X_test, y_test):
    model = clf.fit(X_train, y_train)
    joblib.dump(model, f"{name}.pkl")
    y_proba = model.predict_proba(X_test)[:, 1]
    y_pred = model.predict(X_test)

    # 计算fpr, tpr及相应的thresholds
    fpr, tpr, thresholds = roc_curve(y_test, y_proba)

    # 计算sensitivity/recall/TPR, specificity/TNR, PPV, NPV
    sensitivity = tpr
    specificity = 1 - fpr
    ppv = sensitivity / (sensitivity + fpr+1)
    npv = specificity / (specificity + (1 - tpr))

    # 计算F1 Score
    f1_scores = 2*(ppv*sensitivity) / (ppv + sensitivity)

    # 找到最佳阈值
    ix = argmax(f1_scores)
    best_threshold = thresholds[ix]/4

    # 基于最佳阈值，得到新的预测值
    y_pred_new = [1 if i >= best_threshold else 0 for i in y_proba]

    # 基本度量
    acc = accuracy_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_proba)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)

    # confusion matrix
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()

    # specificity，brier分数，PPV和NPV
    specificity = tn / (tn+fp)
    brier = brier_score_loss(y_test, y_proba)
    ppv = tp / (tp + fp+1) # PPV就是precision
    npv = tn / (tn + fn)

    return (best_threshold, acc, auc, recall, specificity, f1, precision, ppv, npv, brier)

result_train = []
result_test = []

for name, clf in classifiers:
    train_result = evaluate_model(name, clf, X_train, y_train, X_train, y_train)
    test_result = evaluate_model(name, clf, X_train, y_train, X_test, y_test)

    result_train.append((name,) + train_result)
    result_test.append((name,) + test_result)

df_result_train = pd.DataFrame(result_train, columns=['model','best_threshold', 'accuracy', 'AUC', 'Sensitivity(Recall)', 'Specificity', 'F1_Score', 'Precision', 'PPV', 'NPV', 'Brier Score'])
df_result_test = pd.DataFrame(result_test, columns=['model','best_threshold', 'accuracy', 'AUC', 'Sensitivity(Recall)', 'Specificity', 'F1_Score', 'Precision', 'PPV', 'NPV', 'Brier Score'])

df_result_train.to_excel('model_evaluation_result_train.xlsx', index=False)
df_result_test.to_excel('model_evaluation_result_test.xlsx', index=False)
```

```python
def decision_curve(y_true, probs, thresholds, prevalence):
    net_benefits = []
    net_benefit_treat_all = []
    N = len(y_true)
    
    for t in thresholds:
        predicted_positive = probs >= t
        TP = np.sum((predicted_positive == 1) & (y_true == 1))
        FP = np.sum((predicted_positive == 1) & (y_true == 0))
        nb = (TP / N) - (FP / N) * (t / (1 - t))
        net_benefits.append(nb)
        
        # Treat All策略的净收益
        treat_all_nb = prevalence - (1 - prevalence) * (t / (1 - t))
        net_benefit_treat_all.append(treat_all_nb)
        
    return net_benefits, net_benefit_treat_all

# 设定阈值范围
thresholds = np.linspace(0.01, 0.99, 100)

y_true = y_test
probs = y_pred_proba

# 计算
prevalence = np.mean(y_true)  # 正例比例
nb, nb_treat_all = decision_curve(y_true, probs, thresholds, prevalence)

# 绘图
plt.figure(figsize=(10, 6), dpi=300)
plt.plot(thresholds, nb, label='Model', color="black")
plt.plot(thresholds, nb_treat_all, '--', label='Treat All', color="red")
plt.plot(thresholds, np.zeros_like(thresholds), '--', color="orange", label='Treat None')

plt.xlabel('Threshold Probability', fontsize=12)
plt.ylabel('Net Benefit', fontsize=12)
plt.title('Decision Curve Analysis', fontsize=15)
plt.xlim(0, 1)
plt.ylim(-2, 2)
plt.legend(loc="lower center", ncols=3, frameon=False)

plt.tight_layout()
plt.show()
```

```python
from sklearn.calibration import calibration_curve
from sklearn.metrics import brier_score_loss

def brier_score_confidence_interval(y_true, y_probs, n_bootstraps=1000, alpha=0.95):
    brier_scores = []
    rng = np.random.RandomState(seed=42)  # 固定随机种子
    # 将 y_true 和 y_probs 转换为 NumPy 数组，避免 Pandas 索引问题
    y_true = np.array(y_true)
    y_probs = np.array(y_probs)
    
    for _ in range(n_bootstraps):
        indices = np.random.choice(len(y_true), len(y_true), replace=True)
        y_true_resampled = y_true[indices]
        y_probs_resampled = y_probs[indices]
        brier_scores.append(brier_score_loss(y_true_resampled, y_probs_resampled))
    
    brier_mean = np.mean(brier_scores)
    lower_bound = np.percentile(brier_scores, (1 - alpha) / 2 * 100)
    upper_bound = np.percentile(brier_scores, (alpha + (1 - alpha) / 2) * 100)
    
    return brier_mean, lower_bound, upper_bound

# 绘制校准曲线函数
def plot_calibration_curves(models, model_names, X, y, dataset_name):
    plt.figure(figsize=(8, 6))
    plt.plot([0, 1], [0, 1], linestyle='--', label='Perfectly Calibrated')

    for model, name in zip(models, model_names):
        # 获取模型预测概率
        y_pred_prob = model.predict_proba(X)[:, 1]
        
        # 计算校准曲线
        prob_true, prob_pred = calibration_curve(y, y_pred_prob, n_bins=10)
        
        # 计算 Brier 分数和置信区间
        brier_mean, brier_lower, brier_upper = brier_score_confidence_interval(y, y_pred_prob)
        
        # 绘制校准曲线
        plt.plot(prob_pred, prob_true, marker='o', linewidth=1, 
                 label=f'{name} [Brier={brier_mean:.3f} 95%CI ({brier_lower:.3f}-{brier_upper:.3f})]')

    plt.title(f'Calibration Curve ({dataset_name})')
    plt.xlabel('Mean predicted value')
    plt.ylabel('Fraction of positives')
    
    # 将图例固定在右下角
    plt.legend(loc='lower right')
    
    # 保存图像
    plt.savefig(f'calibration_curve_{dataset_name.lower()}.pdf', dpi=1200, bbox_inches='tight', format='pdf')
    plt.show()

models = [model]  
model_names = ['XGBoost']

# 使用训练数据进行校准曲线绘制
plot_calibration_curves(models, model_names, X_test, y_test, "Test Set")
```

```python
# 函数：计算模型的净效益
def calculate_net_benefit_model(thresh_group, y_pred_score, y_label):
    net_benefit_model = np.zeros_like(thresh_group)
    n = len(y_label)
    for i, thresh in enumerate(thresh_group):
        y_pred_label = y_pred_score > thresh
        tn, fp, fn, tp = confusion_matrix(y_label, y_pred_label).ravel()
        net_benefit = (tp / n) - (fp / n) * (thresh / (1 - thresh))
        net_benefit_model[i] = net_benefit
    return net_benefit_model

# 函数：计算Treat All策略的净效益
def calculate_net_benefit_all(thresh_group, y_label):
    net_benefit_all = np.zeros_like(thresh_group)
    n = len(y_label)
    for i, thresh in enumerate(thresh_group):
        predictions = np.ones(n)  # 假设全体都被预测为阳性
        tn, fp, fn, tp = confusion_matrix(y_label, predictions).ravel()
        net_benefit = (tp / n) - (fp / n) * (thresh / (1 - thresh))
        net_benefit_all[i] = net_benefit
    return net_benefit_all

# 函数：绘制DCA曲线
def plot_DCA(ax, thresh_group, net_benefit_model, label, color):
    ax.plot(thresh_group, net_benefit_model, label=label, color=color, linewidth=2)
    ax.set_xlim(0, 1)

    # Check for NaN or Inf values and replace them
    if np.any(np.isnan(net_benefit_model)) or np.any(np.isinf(net_benefit_model)):
        net_benefit_model = np.nan_to_num(net_benefit_model, nan=0.0, posinf=0.0, neginf=0.0)

    ax.set_ylim(net_benefit_model.min() - 0.05, net_benefit_model.max() + 0.05)
    ax.set_xlabel('Threshold Probability', fontsize=12, fontweight='bold', fontfamily='Arial')
    ax.set_ylabel('Net Benefit', fontsize=12, fontweight='bold', fontfamily='Arial')
    ax.grid(True, linestyle='--', alpha=0.7)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.legend(loc='upper right', fontsize=10, frameon=False, ncols=3)
    return ax

# 绘制DCA的主函数
def plot_decision_curve_analysis(models, model_names, thresh_group, X_train, y_train, X_test, y_test, fig_name='dca_plot.png'):
    fig, ax = plt.subplots(figsize=(10, 6))

    # 计算Treat All策略的净效益
    net_benefit_all = calculate_net_benefit_all(thresh_group, y_test)

    # 绘制Treat all和Treat none的基线策略
    ax.plot(thresh_group, net_benefit_all, color='black', linestyle='-', label='Treat all', linewidth=2)
    ax.plot((0, 1), (0, 0), color='gray', linestyle='--', label='Treat none', linewidth=2)

    # 绘制每个模型的净效益曲线
    colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b']
    for i, (label, model) in enumerate(models.items()):
        # 获取每个模型的预测概率
        preds_test = model.predict_proba(X_test)[:, 1]  # 使用模型的测试集预测概率
        net_benefit_model = calculate_net_benefit_model(thresh_group, preds_test, y_test)
        # 使用model_names中的对应名称作为图例
        ax = plot_DCA(ax, thresh_group, net_benefit_model, model_names[i], colors[i])

    # 图像的美观设置
    ax.set_title('Decision Curve Analysis (Test set)', fontsize=16, fontweight='bold', fontfamily='Arial', pad=20)
    ax.tick_params(axis='both', which='major', labelsize=12)

    plt.ylim(-0.2, 0.5)
    
    # 保存图像
    plt.tight_layout()
    plt.savefig(fig_name, dpi=1200, bbox_inches='tight', format='pdf')
    plt.show()

# 将XGBoost模型添加到模型字典中
models = {'XGBoost': model}
model_names = ['XGBoost']
# 设置阈值范围
thresh_group = np.linspace(0, 0.98, 98)
# 使用DCA函数绘制模型在测试集上的DCA曲线
plot_decision_curve_analysis(models, model_names, thresh_group, X_train, y_train, X_test, y_test, fig_name='dca_test_plot.pdf')
```
